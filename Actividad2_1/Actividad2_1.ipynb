```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Datos_Facturaci√≥n.csv")
df
```

```python
#Valores de 1000 a 2000 en CVE_CLPV
filtro=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro
```

```python
max(filtro["CVE_CLPV"])
```

```python
min(filtro["CVE_CLPV"])
```

```python
filtro.to_csv("Ejercicio1.csv")
```

```python
#Se seleccionan todas las claves  excepto la 4 y 5 en CVE_VEND
filtro2=df[~((df["CVE_VEND"] == 4) & (df["CVE_VEND"] == 5))]
filtro2
```

```python
filtro2.to_csv("Ejercicio2.csv")
```

```python
#Se selecciona la fecha 28/02/2022 en fecha_ent
filtro3=df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtro3
```

```python
filtro3.to_csv("Ejercicio3.csv")
```

```python
#Cantidades menores a 5951.7 o status E en CANT_TOT
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
filtro4.to_csv("Ejercicio4.csv")
```

```python
#Se seleccionan las columnas CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5=df.iloc[ : , [1,7,8,10]]
filtro5
```

```python
filtro5.to_csv("Ejercicio5.csv")
```
